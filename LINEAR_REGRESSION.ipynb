### TRY LINIEAR REGRESSION BUT THE CORRELATION NOT PRETY 
### GOOD WITH PRICE AND BEDROOMS IS 0.308350

```python
fig = plt.figure(figsize=(10, 10))

nilaix = df['bedrooms']
nilaiy= df['price']

slope, intercept, r, p, std_err = stats.linregress(nilaix, nilaiy)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, nilaix))

plt.xlabel("Open price")
plt.ylabel("Close Price")
plt.scatter(nilaix,nilaiy)
plt.plot(nilaix,mymodel,color='r')
plt.show()
```

### Bisa dilihat gambar di atas karena correlation tidak cukup baik maka gambar visual seperti di atas

```python
# kita coba membuat split data train dan data test dengan menggunakan train test 40% dengan test 60%

# Reshape the 'price' feature
X = df['bedrooms'].values.reshape(-1, 1)
y = df['price']

# Split the data into training and testing sets
X_train, test_X, y_train, y_test = train_test_split(X, y, train_size=0.4, shuffle=False)

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(test_X)
```

```python
intercept =model.intercept_
coefficients = model.coef_
```

```python
# Setelah model dilatih
coef = model.coef_
intercept = model.intercept_

# Menampilkan Intercept
print("Intercept:", intercept)

# Menampilkan Koefisien untuk Fitur Tunggal
print("Coefficient for 'price':", coefficients)
```

```python
## evaluation with RMSE and MAE
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_test, predictions, squared=False)
print("Root Mean Squared Error:", rmse)
```

```python
print("mean absolute error", mean_absolute_error(y_test,predictions))
```
